```python
#A1

import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load the audio file 'ruchith.wav' and its sampling rate
y, sr = librosa.load('ruchith.wav')

# Set up the plot
plt.figure(figsize=(12, 6))

# Plot the waveform of the audio signal
librosa.display.waveshow(y, sr=sr, alpha=0.8, color='r')

# Add title and axis labels
plt.title('Speech Signal Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

# Add grid lines
plt.grid(True)

# Adjust layout for better appearance
plt.tight_layout()

# Display the plot
plt.show()


```
